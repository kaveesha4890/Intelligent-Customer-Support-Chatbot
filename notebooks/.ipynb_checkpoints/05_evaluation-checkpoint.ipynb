# 05 - Full System Evaluation
Runs the chatbot pipeline on the test query set and computes evaluation metrics + visualizations.

```python
import sys, os
sys.path.append(os.path.abspath(".."))

import pandas as pd
import matplotlib.pyplot as plt
from src.chatbot_pipeline import chat
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu
import nltk
nltk.download("punkt")
```

```python
test_set = pd.read_csv("../evaluation/test_queries.csv")
scorer = rouge_scorer.RougeScorer(["rougeL"], use_stemmer=True)

results = []
for _, row in test_set.iterrows():
    response, intent, sentiment, esc = chat(row["query"], [])
    bleu = sentence_bleu([row["reference"].split()], response.split())
    rouge_l = scorer.score(row["reference"], response)["rougeL"].fmeasure
    results.append({
        "query": row["query"],
        "reference": row["reference"],
        "response": response,
        "bleu": bleu,
        "rouge_l": rouge_l,
        "intent": intent,
        "sentiment": sentiment,
        "escalated": esc
    })

results_df = pd.DataFrame(results)
results_df.to_csv("../evaluation/results.csv", index=False)
results_df.head()
```

```python
print("Average BLEU:", results_df["bleu"].mean())
print("Average ROUGE-L:", results_df["rouge_l"].mean())
print("Escalation rate:", results_df["escalated"].mean())
```

```python
plt.figure(figsize=(8,4))
results_df[["bleu", "rouge_l"]].mean().plot(kind="bar")
plt.title("Average BLEU and ROUGE-L across test set")
plt.ylabel("Score")
plt.show()
```

```python
plt.figure(figsize=(8,4))
results_df["intent"].value_counts().head(15).plot(kind="bar")
plt.title("Predicted Intent Distribution (Test Set)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
results_df["sentiment"].value_counts().plot(kind="bar", color=["green","gray","orange","red"])
plt.title("Sentiment Distribution (Test Set)")
plt.ylabel("Count")
plt.show()
```
